```python
#3번

name = input ('이름을 입력해주세요 ')
grade = input ('점수를 입력해주세요 ')

def grad (a):
  if 95 <= a <= 100:
    return 'A+'
  elif 90 <= a <= 94:
    return 'A'
  elif 85 <= a <= 89:
    return 'B+'
  elif 80 <= a <= 84:
    return 'B'
  elif 75 <= a <= 79:
    return 'C+'
  elif 70 <= a <= 74:
    return 'C'
  elif 65 <= a <= 69:
    return 'D+'
  elif 60 <= a <= 64:
    return 'D'
  else:
    return 'F'

print('\n학생이름:', name)
print('점수:', grade)
print('학점:', grad (int(grade)))
```
